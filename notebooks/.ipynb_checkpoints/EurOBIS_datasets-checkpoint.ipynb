#### Types of datasets:

```python
#helper functions 
def get_info_simple(df: pd.DataFrame, key:str):
    if key in df.keys():
        result = df.groupby(key, dropna=False).nunique()
    else:
        result = f'No {key} information available'
        
    return result

def visualize_missing_info(df: pd.DataFrame, column_name:str) -> pd.DataFrame:
    nulls=pd.DataFrame(df.isnull().sum())
    notnulls=pd.DataFrame(df.notnull().sum())
    missing_info = nulls.merge(notnulls, left_index=True, right_index=True)
    missing_info.columns = ['missing','present']
    
    missing_info_t = missing_info.transpose()

    plt.pie(missing_info_t[column_name], labels=missing_info_t.index, autopct='%1.1f%%')
    plt.show()
    return missing_info_t
```

```python
types = execute_to_df("dataset_type.sparql")
```

```python
get_info_simple(types, 'type')
```
