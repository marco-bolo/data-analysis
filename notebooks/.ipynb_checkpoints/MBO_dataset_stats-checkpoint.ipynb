- assessment of what url points to (is this machine accessible? useable format?, doubles?, ...)
- in case of json record --> conversion of json into json-ld using bioships template, output stored in /output

```python
def get_missing_info(df: pd.DataFrame, col: str) -> pd.DataFrame:
    missing_info = pd.DataFrame([[df[col].count(), df[col].isnull().sum()]], columns=['Present', 'Missing'], index=[col])

    missing_info.loc[col].plot.pie(autopct='%1.1f%%', startangle=140)
    plt.title(f'Missing vs Non-Missing Values in {col}')
    plt.ylabel('')  # Remove y-axis label for better appearance
    return missing_info


def get_info(df: pd.DataFrame, col: str) -> pd.DataFrame:
    distinct_values = df[col].unique()
    data = [(df[col] == distinct_values[i]).sum() for i,val in enumerate(distinct_values)]
    info = pd.DataFrame([data], columns=distinct_values, index=[col])

    info.loc[col].plot.pie(autopct='%1.1f%%')
    plt.title(f'{distinct_values} Values in {col}')
    plt.ylabel('')  # Remove y-axis label for better appearance
    return info
    
```

### WP3 datasets

```python
WP3 = pd.read_csv("/data/MBO_WP3_description_added.csv")
```

```python
WP3.info()
```

```python
# info on DataLandingPageURL
WP3.DataLandingPageURL.describe()
result = get_missing_info(WP3, 'DataLandingPageURL')
```

```python
# info on DataLandingPageURL_result 
result = get_info(WP3, 'DataLandingPageURL_result')
```

```python
# DataLandingPageURL_result_MachineAccessible
result = get_info(WP3, 'DataLandingPageURL_result_MachineAccessible')
```

Explanation of scores:  
0 = fully machine accessible, url directly results in desired document   
1 = url results in desired document but requires small manipulation (e.g. adding suffix)  
2 = url results in desired document but requires manipulation with knowledge of internal data model (e.g. knowing which number in url is dasid and what a dasid represents)  
3 = url results in document but not in desired format and requires further knowledge on internal data model to understand  
4 = not machine accessible, doesn't result in document   

```python
# info on DatasetDouble
result = get_info(WP3, 'DatasetDouble')
```

```python
# info on DatasetConversion
result = get_info(WP3, 'DatasetConversion')
```
